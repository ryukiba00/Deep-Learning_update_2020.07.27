```
class Dog(object):
  def __init__(self, breed):
    self.breed = breed

sam = Dog(breed = "Lab")
frank = Dog(breed = "Huskie")

print(sam.breed)
print(frank.breed)
```

```
class Dog(object):
  species = "mammal"
  def __init__(self, breed, name):
    self.breed = breed
    self.name = name

sam = Dog("Lab","Sam")
print(sam.name)

print(Dog.species)
```

```
class Circle(object):
  pi = 3.14
  def __init__(self, radius = 1):
    self.radius = radius
  def area(self):
    return self.radius * self.radius * Circle.pi
  def setRadius(self, radius):
    self.radius = radius
  def getRadius(self):
    return self.radius
```

```
c = Circle()
c.area()
```

```
c.setRadius(2)
```

```
c.getRadius()
```
